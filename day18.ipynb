```python
import os
from collections import defaultdict

import sys
sys.setrecursionlimit(5000)

cwd = os.getcwd()
file = cwd + '/inputs/input18.txt'

data = open(file).read().splitlines()

max_pixel_coord = 20

pixel_cube = []
for x in range(max_pixel_coord):
  pixel_matrix = []
  for y in range(max_pixel_coord):
    pixel_row = [0] * max_pixel_coord
    pixel_matrix.append(pixel_row)
  pixel_cube.append(pixel_matrix)

for i in range(len(data)):
  pixel_list = data[i].split(',')
  pixel_cube[int(pixel_list[0])][int(pixel_list[1])][int(pixel_list[2])] = 1

# for i in range(len(pixel_cube[0])):
#   print('Next')
#   for j in range(len(pixel_cube[0][0])):
#     print(pixel_cube[i][j])

# spread steam

def steam_check(x,y,z):
  if pixel_cube[x][y][z] > 0:
    return
  pixel_cube[x][y][z] = 2
  if x - 1 >= 0:
    steam_check(x - 1, y, z)
  if x + 1 < max_pixel_coord:
    steam_check(x + 1, y, z)
  if y - 1 >= 0:
    steam_check(x, y - 1, z)
  if y + 1 < max_pixel_coord:
    steam_check(x, y + 1, z)
  if z - 1 >= 0:
    steam_check(x, y, z - 1)
  if z + 1 < max_pixel_coord:
    steam_check(x, y, z + 1)

for x in range(max_pixel_coord):
  steam_check(x, 0, 0)
for x in range(max_pixel_coord):
  steam_check(x, max_pixel_coord - 1, max_pixel_coord - 1)

for y in range(max_pixel_coord):
  steam_check(0, y, 0)
for y in range(max_pixel_coord):
  steam_check(max_pixel_coord - 1, y, max_pixel_coord - 1)

for z in range(max_pixel_coord):
  steam_check(0, 0, z)
for z in range(max_pixel_coord):
  steam_check(max_pixel_coord - 1, max_pixel_coord - 1, z)

area = len(data) * 6

# z axis check
for x in range(max_pixel_coord):
  for y in range(max_pixel_coord):
    previous_hit = False
    previous_air = False
    for z in range(max_pixel_coord):
      if pixel_cube[x][y][z] == 1:
        if previous_hit:
          area -= 2
        if previous_air:
          area -= 1
        previous_hit = True
        previous_air = False
      elif pixel_cube[x][y][z] == 0:
        if previous_hit:
          area -= 1
        previous_hit = False
        previous_air = True
      else:
        previous_hit = False
        previous_air = False

# y axis check
for x in range(max_pixel_coord):
  for z in range(max_pixel_coord):
    previous_hit = False
    previous_air = False
    for y in range(max_pixel_coord):
      if pixel_cube[x][y][z] == 1:
        if previous_hit:
          area -= 2
        if previous_air:
          area -= 1
        previous_hit = True
        previous_air = False
      elif pixel_cube[x][y][z] == 0:
        if previous_hit:
          area -= 1
        previous_hit = False
        previous_air = True
      else:
        previous_hit = False
        previous_air = False

# x axis check
for y in range(max_pixel_coord):
  for z in range(max_pixel_coord):
    previous_hit = False
    previous_air = False
    for x in range(max_pixel_coord):
      if pixel_cube[x][y][z] == 1:
        if previous_hit:
          area -= 2
        if previous_air:
          area -= 1
        previous_hit = True
        previous_air = False
      elif pixel_cube[x][y][z] == 0:
        if previous_hit:
          area -= 1
        previous_hit = False
        previous_air = True
      else:
        previous_hit = False
        previous_air = False

print(area)

sys.setrecursionlimit(1000)
```
