```python
import os

cwd = os.getcwd()
file = cwd + '/inputs/input09.txt'

data = open(file).read().splitlines()

head_position = [0,0]
tail_position = [0,0]
path_matrix = [[1]]

def adjust_tail():
  if abs(head_position[0] - tail_position[0]) >= 2:
    if head_position[0] < tail_position[0]:
      tail_position[0] += -1
    else:
      tail_position[0] += 1
    if head_position[1] < tail_position[1]:
      tail_position[1] += -1
    elif head_position[1] > tail_position[1]:
      tail_position[1] += 1
  elif abs(head_position[1] - tail_position[1]) >= 2:
    if head_position[1] < tail_position[1]:
      tail_position[1] += -1
    else:
      tail_position[1] += 1
    if head_position[0] < tail_position[0]:
      tail_position[0] += -1
    elif head_position[0] > tail_position[0]:
      tail_position[0] += 1
  path_matrix[tail_position[1]][tail_position[0]] = 1


def move_left(distance):
  destination = head_position[0] - distance
  if destination < 0:
    overflow = abs(destination)
    head_position[0] = head_position[0] + overflow
    tail_position[0] = tail_position[0] + overflow
    for i in range(len(path_matrix)):
      new_space = [0] * overflow
      path_matrix[i] = new_space + path_matrix[i]
  for x in range(distance):
    head_position[0] += -1
    adjust_tail()

def move_up(distance):
  destination = head_position[1] - distance
  if destination < 0:
    overflow = abs(destination)
    head_position[1] = head_position[1] + overflow
    tail_position[1] = tail_position[1] + overflow
    for i in range(overflow):
      new_space = [0] * len(path_matrix[0])
      path_matrix.insert(0, new_space)
  for x in range(distance):
    head_position[1] += -1
    adjust_tail()

def move_right(distance):
  destination = head_position[0] + distance
  if destination >= len(path_matrix[0]):
    overflow = destination - len(path_matrix[0]) + 1
    for i in range(len(path_matrix)):
      new_space = [0] * overflow
      path_matrix[i] = path_matrix[i] + new_space
  for x in range(distance):
    head_position[0] += 1
    adjust_tail()

def move_down(distance):
  destination = head_position[1] + distance
  if destination >= len(path_matrix):
    overflow = destination - len(path_matrix) + 1
    for i in range(overflow):
      new_space = [0] * len(path_matrix[0])
      path_matrix.append(new_space)
  for x in range(distance):
    head_position[1] += 1
    adjust_tail()

for line in data:
  steps = line.strip().split(" ")

  direction = steps[0]
  if steps[0] == 'L':
    move_left(int(steps[1]))
  elif steps[0] == 'U':
    move_up(int(steps[1]))
  elif steps[0] == 'R':
    move_right(int(steps[1]))
  elif steps[0] == 'D':
    move_down(int(steps[1]))

solution = 0

for row in path_matrix:
  for entry in row:
    solution += entry

print(solution)
```
