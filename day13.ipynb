```python
import os
import functools
from collections import defaultdict

cwd = os.getcwd()
file = cwd + '/inputs/input13.txt'

data = open(file).read().splitlines()

def convert_string_to_list(string, i=0):
  content = ''
  conversion = []
  while i < len(string):
    i += 1
    c = string[i]
    if c == ']' or c ==',':
      if content != '':
        if not isinstance(content, list):
          content = int(content)
        conversion.append(content)
        content = ''
      if c == ']':
        return {"conversion": conversion, "index": i}
    elif c == '[':
      result = convert_string_to_list(string, i)
      content = result["conversion"]
      i = result["index"]
    else:
      content += c
    
def compare_packets(p1,p2):
  p1_is_list = isinstance(p1, list)
  p2_is_list = isinstance(p2, list)
  if not p1_is_list and not p2_is_list:
    if p1 < p2:
      return 'correct'
    elif p1 > p2:
      return 'incorrect'
    else:
      return 'continue'
  elif p1_is_list and p2_is_list:
    for i in range(len(p1)):
      if i < len(p2):
        comparison = compare_packets(p1[i],p2[i])
        if comparison != 'continue':
          return comparison
      else:
        return 'incorrect'
    if len(p1) == len(p2):
      return 'continue'
    else:
      return 'correct'
  else:
    if p1_is_list:
      p2 = [p2]
    else:
      p1 = [p1]
    return compare_packets(p1,p2)

def sort_packets(p1,p2):
  result = compare_packets(p1,p2)
  if result == 'correct':
    return -1
  elif result == 'incorrect':
    return 1
  else:
    return 0

converted_data = [[[2]],[[6]]]
for line in data:
  if len(line) > 0:
    converted_data.append(convert_string_to_list(line)["conversion"])

converted_data.sort(key=functools.cmp_to_key(sort_packets))

print(((1 + converted_data.index([[2]])) * (1 + converted_data.index([[6]]))))
```
